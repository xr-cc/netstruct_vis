# Demo of data visualization on folklore motif clustering results

```python
import netstruct_vis as ns_vis
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataPath = "data/"
```

**Extract hierarhical tree from NetStruct output files**

```python
belonging_map, parent_map, child_map, node_list, G_whole = ns_vis.extractTree(dataPath+"NS_output_demo_folklore/")
```

**Load meta information on individuals**

```python
df_meta = pd.read_csv(dataPath+"sample_data_folklore.txt")
df_meta.head()
```

**Plot the complete tree using the gradient coloring scheme and corresponding individuals on map**

```python
cmap = plt.cm.rainbow
fig_tree, G, color_int_map = ns_vis.plotSubTree(child_map, belonging_map, cmap, root='0-0-0', 
                                                node_size=300, figsize=(5,5), dpi=70)
fig_tree.savefig('plots/folklore_tree.png',dpi=70, bbox_inches = 'tight',transparent=True) 
```

```python
fig_map = ns_vis.plotMap(df_meta, belonging_map, color_int_map, G, (-60, 75, -175, 175), cmap, 
                  lat_lb='latitude', lon_lb='longitude', pt_size=40, annot_lb=None, figsize=(9, 6), dpi=70)
fig_map.savefig('plots/folklore_map.png',dpi=70, bbox_inches = 'tight',transparent=True) 
```

**Plot the left subtree using the gradient coloring scheme and corresponding individuals on map**

```python
cmap = plt.cm.rainbow
fig_test, G, color_int_map = ns_vis.plotSubTree(child_map, belonging_map, cmap, root='1-0-0', 
                                                node_size=300, figsize=(5,4), dpi=70)
```
